## Calculate cdists

### TCRdist

```python
tr = TCRrep(cell_df=df, organism="human", chains=["beta"])
tcrdist_pdist = squareform(tr.pw_beta)
df = tr.clone_df
```

```python
epitope = df["epitope"].to_numpy()
within = (epitope[:, np.newaxis] == epitope[np.newaxis, :]) & (
    ~(np.eye(len(epitope)) == 1)
)
within = squareform(within)
```

### BCDRBERT

```python
if torch.cuda.is_available():
    device = torch.device("cuda:0")
else:
    device = torch.device("cpu")
```

```python
SAVE_DIR = Path("/home/yutanagano/UCLOneDrive/MBPhD/projects/tcr_embedder/model_saves/")
MODEL_DIR = SAVE_DIR / "BCDRBERT_+ACL_cdrcns_020"

model = ModelLoader(MODEL_DIR)

# model = modules.AtchleyEmbedder()
# model.to(device)
```

```python
bvcdr3df = df.rename(columns={"v_b_gene": "TRBV", "cdr3_b_aa": "CDR3B"})
```

```python
bvcdr3bert_pdist = model.pdist(bvcdr3df)
```

```python
fig, ax = plt.subplots()
PrecisionRecallDisplay.from_predictions(
    within, np.exp(-tcrdist_pdist / 50), ax=ax, name="TCRdist"
)
PrecisionRecallDisplay.from_predictions(
    within, np.exp(-bvcdr3bert_pdist / 50), ax=ax, name=model.name
)
```
